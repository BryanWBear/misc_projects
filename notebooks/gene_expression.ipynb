```python
PURE_COLS = [col for col in df.columns if '100' in col]

BRAIN_COLS = [col for col in PURE_COLS if 'B' in col]
LIVER_COLS = [col for col in PURE_COLS if 'L' in col]
PLACENTA_COLS = [col for col in PURE_COLS if 'P' in col]


def get_basis_vecs(df): 
    brain_vec = df[BRAIN_COLS].mean(axis=1)
    liver_vec = df[LIVER_COLS].mean(axis=1)
    pla_vec = df[PLACENTA_COLS].mean(axis=1)
    return brain_vec, liver_vec, pla_vec

brain_vec, liver_vec, pla_vec = get_basis_vecs(df)
bv_filt, lv_filt, pv_filt = get_basis_vecs(df_filt)
```

```python
from numpy.linalg import inv
import numpy as np

# project w onto u and v
def non_orthogonal_projection(u, v, w):
    uw = np.dot(u, w)
    uv = np.dot(u, v)
    vw = np.dot(v, w)
    uu = np.dot(u, u)
    vv = np.dot(v, v)

    proj = np.array([[uu, uv], [uv, vv]])

    return inv(proj) @ np.array([uw, vw])



def get_projections(df, brain_vec, liver_vec):
    print(non_orthogonal_projection(brain_vec, liver_vec, df['X80B_20L_1']))
    print(non_orthogonal_projection(brain_vec, liver_vec, df['X60B_40L_1']))
    print(non_orthogonal_projection(brain_vec, liver_vec, df['X40B_60L_1']))
    print(non_orthogonal_projection(brain_vec, liver_vec, df['X20B_80L_1']))

get_projections(df_filt, bv_filt, lv_filt)
```

```python
from sklearn.svm import SVC

clf = SVC(kernel='linear', gamma=0.01, probability=True)
clf.fit(X_train, y_train)
```
